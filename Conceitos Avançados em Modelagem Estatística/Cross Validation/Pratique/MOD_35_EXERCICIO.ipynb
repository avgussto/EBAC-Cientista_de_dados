E a variável alvo:

O objetivo desta tarefa é aplicar a técnica de validação cruzada (cross-validation) para avaliar a performance de um modelo de classificação. A validação cruzada ajudará a garantir que o modelo seja avaliado de maneira robusta e generalize bem para dados não vistos.

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
```

# 1 - Carregue a base de dados, verifique os tipos de dados e também se há presença de dados faltantes ou nulos.

```python
df = pd.read_csv('Cientista de dados M35 - smoke_detection_iot.csv')

df.drop(columns=['Unnamed: 0'], inplace=True)
df.rename(columns={'Fire Alarm': 'Fire_Alarm'}, inplace=True)

df.head()
```

# 2 - Para essa base, onde você realizará as previsões de fire alarm, qual modelo de machine learning você aplicará? Justifique.




Optei por utilizar o Random Forest, sua robustez e prevenção a overfitting me chamam a atenção. # 3 - Separe a base em Y e X e já rode a instância do modelo que você utilizará.

```python
# separando a variável target
X = df.drop(columns=['Fire_Alarm'])
y = df['Fire_Alarm']

# importando e instanciando o random forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=42)
```

# 4 - Defina o número de Folds e rode o modelo com a validação cruzada.

```python
# definindo o kfold
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# realizando o cross validation
scores = cross_val_score(rf, X, y, cv=kf, scoring='accuracy')
print(f'Scores de cada fold: {scores}\n')
print(f'Accuracy médio: {scores.mean()}')
```

# 5 - Avalie a pontuação de cada modelo e ao final a validação final da média.

```python
# scores de cada modelo
print(f'Scores de cada fold: {scores}')

# accuracy médio
print(f'Accuracy médio: {scores.mean()}')
```

A acurácia veio extremamente próxima de 1, existe a possibildade do modelo ter ficado extremamente eficiente, mas uma acurácia tão alta é um forte indício de overfitting. Estudar o balanceamento das classes e correlação das variáveis são ações recomendadas para compreender melhor os resultados do modelo
