```python
import pandas as pd
import random
from faker import Faker
```

```python
faker = Faker('pt_BR')

dados_pessoais = []

for _ in range(100):
    dados_pessoais.append({
        'nome': faker.name(),
        'email': faker.email(),
        'cpf': faker.cpf(),
        'data_nascimento': faker.date_of_birth(minimum_age=18, maximum_age=60).strftime('%d, %m, %Y'),
        'idade': random.randint(18, 60),
        'endereco': faker.address(),
        'cidade': faker.city(),
        'estado': faker.state_abbr(),
        'telefone': faker.phone_number()
    })

df_pessoas = pd.DataFrame(dados_pessoais)

df_pessoas

df_pessoas.to_csv('pessoas.csv', index=False)
```
