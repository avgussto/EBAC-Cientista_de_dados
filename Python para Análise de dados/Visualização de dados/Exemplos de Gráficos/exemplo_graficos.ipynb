```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('clientes-v3-preparado.csv')

df.head()
```

```python
# Heatmap de correlação 
df_numeric = df.select_dtypes(include=['number'])

plt.figure(figsize=(10, 10))
sns.heatmap(df_numeric.corr(), annot=True, fmt='.2f')
plt.title('Heatmap de correlação entre as variáveis')
plt.show()
```

```python
# Countplot 
sns.countplot(x='estado_civil', data=df)
plt.title('Distribuição de clientes por estado civil')
plt.xlabel('Estado civil')
plt.ylabel('Quantidade de clientes')
plt.show()
```

```python
# Countplot com legenda
plt.figure(figsize=(10, 6))
sns.countplot(x='estado_civil', data=df, hue='nivel_educacao',palette='viridis')  
plt.title('Distribuição de clientes por estado civil e sexo')
plt.legend(title='Nível de educação', loc='upper right')
plt.xlabel('Estado civil')
plt.ylabel('Quantidade de clientes')
```
