```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('clientes-v3-preparado.csv')
df.head()
```

```python
# grafico de barras
plt.figure(figsize=(10, 5)) 
df['nivel_educacao'].value_counts().plot(kind='bar', color='purple')
plt.title('Divisão de Escolaridade')
plt.xlabel('Escolaridade')  
plt.ylabel('Quantidade')
plt.xticks(rotation=0)  
plt.show()
```

```python
x = df['nivel_educacao'].value_counts().index
y = df['nivel_educacao'].value_counts().values

plt.figure(figsize=(10, 5))
plt.bar(x, y, color='purple')   
plt.title('Divisão de Escolaridade')
plt.xlabel('Escolaridade') 
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.show()
```

```python
# grafico de pizza - não muito recomendado
plt.figure(figsize=(10, 5))
plt.pie(y, labels = x, autopct= '%.1f%%', startangle=90)
plt.title('Distribuição de Escolaridade')
plt.show()
```

```python
# Grafico de dispersao
plt.figure(figsize=(10, 5))
plt.hexbin(df['idade'], df['salario'], gridsize=40, cmap='Purples')
plt.colorbar(label='Densidade')
plt.xlabel('Salário')
plt.ylabel('Idade')
plt.title('Dispersão de Idade x Salário')
plt.show()
```
