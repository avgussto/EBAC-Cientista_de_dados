```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('clientes-v3-preparado.csv')

df.head()
```

```python
# grafico de dispersão
sns.jointplot(x='idade', y='salario', data=df, kind='scatter')
plt.show()
```

```python
# Grafico de densidade
plt.figure(figsize=(10, 6))
sns.kdeplot(df['salario'], fill=True, color='blue')
plt.title('Densidade de Salário')
plt.xlabel('Salário')   
plt.show()
```

```python
# grafico de pairplot  - dispersao e histograma 
sns.pairplot(df[['idade','anos_experiencia','salario','nivel_educacao']])
plt.show()
```

```python
# grafico de regressao
sns.regplot(x='idade', y='salario', data=df, color='blue',scatter_kws={'alpha':0.5, 'color':'gray'})
plt.title('Regressão Linear Idade x Salário')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.show()
```

```python
# Grafico countplot com hue
sns.countplot(x='estado_civil', data=df, hue='nivel_educacao',palette='viridis')
plt.xlabel('Estado Civil')
plt.ylabel('Total')
plt.title('Total de Clientes por Estado Civil')
plt.legend(title='Nível de Educação')
plt.show()
```
