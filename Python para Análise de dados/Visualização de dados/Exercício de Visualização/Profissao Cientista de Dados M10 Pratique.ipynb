```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('ecommerce_preparados.csv')

df.head()
```

```python
df.columns
```

```python
# Histograma de Preço
plt.figure(figsize=(12,6))
sns.histplot(df['Preço'], bins=30, color='blue', alpha=0.7)
```

```python
# Grafico de dispersão entre preco e quantidade vendida
plt.figure(figsize=(10,6))
sns.scatterplot(x='Preço', y='N_Avaliações', data=df, color='purple')
plt.xlabel('Preço')
plt.ylabel('Número de Avaliações')
plt.title('Preço x Número de Avaliações')
plt.show()
```

```python
df.columns
```

```python
# Mapa de calor
colunas_corr = df[['Preço', 'N_Avaliações', 'Nota', 'Desconto']]
plt.figure(figsize=(10,6))
sns.heatmap(colunas_corr.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlação entre as Variáveis')
plt.show()
```

```python
# Grafico de barras
plt.figure(figsize=(10,6))
df['Marca'].value_counts().head(10).plot(kind='bar', color='green')
```

```python
norm_genero = {'Sem gênero infantil': 'Sem gênero', 'Meninas': 'Feminino', 'Bebês':'Sem gênero', 'Meninos': 'Masculino', 'nan':'Não informado', 'Unissex':'Sem gênero',
               'short menina verao look mulher':'Feminino', 'Mulher':'Feminino', 'roupa para gordinha pluss P ao 52': 'Feminino', 'menino':'Masculino',
               'bermuda feminina brilho Blogueira':'Feminino'}

df['Gênero'] = df['Gênero'].map(norm_genero)
```

```python
# Grafico de pizza
plt.figure(figsize=(10,6))
df['Gênero'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'pink', 'gray'])
```

```python
# Grafico de densidade
plt.figure(figsize=(10,6))
sns.kdeplot(df['Preço'], fill=True, color='red')
plt.title('Densidade de Preço')
plt.ylabel('Densidade')
plt.xlabel('Preço')
plt.show()

```

```python
# Grafico de dispersão com linha de regressão
plt.figure(figsize=(8,5))
sns.regplot(x='Preço_MinMax', y='Nota_MinMax', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.xlabel('Preço do Produto')
plt.ylabel('Notas')
plt.title('Relação entre Nota e Quantidade Vendida')
plt.show()
```
