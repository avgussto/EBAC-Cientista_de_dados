```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('clientes_tratados.csv')
```

```python
df.head()
```

```python
df = pd.concat([df, pd.get_dummies(df['nivel_educacao'], prefix='educacao')], axis=1)
df.head()
```

```python
# codificação ordinal para nivel_educacional

educacao_ordem = {'Ensino Fundamental': 1, 'Ensino Médio':2, 'Ensino Superior':3, 'Pós-Graduação':4, 'Não informado':5}
df['nivel_educacao_ordinal'] = df['nivel_educacao'].map(educacao_ordem)
```

```python
df.head()
```

```python
# usando o metódo cat.codes para codificar a variável categórica area_atuacao

df['area_atuacao'] = df['area_atuacao'].astype('category')  # Converte corretamente
df['area_atuacao_cat'] = df['area_atuacao'].cat.codes       # Agora pode acessar .cat.codes

df.head()
```

```python
# usando o LabelEncoder para codificar a variável categórica estado
le = LabelEncoder()
df['estado_le'] = le.fit_transform(df['estado'])

df.head()
```
