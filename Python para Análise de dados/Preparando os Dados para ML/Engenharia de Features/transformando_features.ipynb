```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('clientes_tratados.csv')
df.head()
```

```python
# Transformação logarítmica
df['salario_log'] = np.log1p(df['salario'])

df[['salario', 'salario_log']].head()
```

```python
# Transformação de box-cox
df['salario_boxcox'], param = stats.boxcox(df['salario'] + 1)

df[['salario', 'salario_boxcox']].head()
```

```python
# frequencia dos estados
df['estado_freq'] = df['estado'].map(df['estado'].value_counts() / len(df))

df[['estado', 'estado_freq']].head()
```

```python
# Interações entre variáveis

df['Interacao_idade_filhos'] = df['idade'] * df['numero_filhos']

df[['idade', 'numero_filhos', 'Interacao_idade_filhos']].head()
```
