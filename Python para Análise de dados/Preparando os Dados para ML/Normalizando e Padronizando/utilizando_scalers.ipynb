```python
import pandas as pd
from sklearn.preprocessing import RobustScaler, StandardScaler, MinMaxScaler, MaxAbsScaler, Normalizer
```

```python
df = pd.read_csv('clientes_tratados.csv')
df.head()
```

```python
df_escalado = df[['idade', 'salario']].copy()
```

```python
min_scaler = MinMaxScaler()

df_escalado['idadeMinMax'] = min_scaler.fit_transform(df_escalado[['idade']])
df_escalado['salarioMinMax'] = min_scaler.fit_transform(df_escalado[['salario']])

robust_scaler = RobustScaler()

df_escalado['idadeRobust'] = robust_scaler.fit_transform(df_escalado[['idade']])
df_escalado['salarioRobust'] = robust_scaler.fit_transform(df_escalado[['salario']])

standard_scaler = StandardScaler()

df_escalado['idadeStandard'] = standard_scaler.fit_transform(df_escalado[['idade']])
df_escalado['salarioStandard'] = standard_scaler.fit_transform(df_escalado[['salario']])

max_abs_scaler = MaxAbsScaler()

df_escalado['idadeMaxAbs'] = max_abs_scaler.fit_transform(df_escalado[['idade']])
df_escalado['salarioMaxAbs'] = max_abs_scaler.fit_transform(df_escalado[['salario']])

normalizer = Normalizer()

df_escalado['idadeNormalizer'] = normalizer.fit_transform(df_escalado[['idade']])
df_escalado['salarioNormalizer'] = normalizer.fit_transform(df_escalado[['salario']])
```
