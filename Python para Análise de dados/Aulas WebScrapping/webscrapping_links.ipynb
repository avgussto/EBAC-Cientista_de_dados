```python
import requests
from bs4 import BeautifulSoup

link = 'https://wiki.python.org.br/AprendaMais'

response = requests.get(link)

extracao = BeautifulSoup(response.text, features='html.parser')
```

```python
for titulo in extracao.find_all('h2'):
    print('\nTitulo: ', titulo.text.strip())    
    for link in titulo.find_next_siblings('p'):
        for a in link.find_all('a', href=True):
            print('Referência: ', a.text.strip(), '| URL', a['href'])
```
