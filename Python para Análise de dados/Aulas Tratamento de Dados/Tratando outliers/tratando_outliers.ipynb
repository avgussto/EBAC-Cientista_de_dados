```python
df = pd.read_csv('clientes_tratados.csv')
df.head()
```

```python
# Observando os outliers em idade
df_filtro_outliers_simples = df[(df['idade'] < 0) | (df['idade'] > 100)]
df_filtro_outliers_simples
```

```python
# Usando o z-score para identificar os outliers
z_scores = stats.zscore(df['idade'])
outliers_z = df[z_scores >= 3]
outliers_z
```

```python
# Identificando outilers com IQR
Q1 = df['idade'].quantile(0.25)
Q3 = df['idade'].quantile(0.75)
IQR = Q3 - Q1

limite_menor = Q1 - 1.5 * IQR
limite_maior = Q3 + 1.5 * IQR

print('Limites do IQR:', limite_menor, limite_maior)
```

```python
outliers_iqr = df[(df['idade'] < limite_menor) | (df['idade'] > limite_maior)]
outliers_iqr
```

```python
df = df.drop(outliers_iqr.index)
```

```python
df.to_csv('clientes_sem_outliers.csv', index=False)
```
